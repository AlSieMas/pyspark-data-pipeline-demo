```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("PySpark Data Pipeline Demo")
    .getOrCreate()
)

spark.version
```

```python
data = [
    ("device-001", 24.1, 0.52),
    ("device-002", 23.8, 0.48),
    ("device-003", 25.0, 0.61),
]

columns = ["device_id", "voltage_v", "current_a"]

df = spark.createDataFrame(data, columns)

df.show()
```

```python
from pyspark.sql import functions as F

df_transformed = df.withColumn(
    "power_w",
    F.round(F.col("voltage_v") * F.col("current_a"),2)
)

df_transformed.show()
```

```python
df_transformed.write.mode("overwrite").parquet("../data/curated/infrastructure_test")
```

```python
df_read_back = spark.read.parquet("../data/curated/infrastructure_test")

df_read_back.show()
```
